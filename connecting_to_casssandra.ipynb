```python
from cassandra.cluster import Cluster
import pandas as pd
```

```python
data = pd.read_csv('/home/tee/BIG_DATA/Air_quality_official.csv')
print(data.columns)
```

```python
# Kết nối đến Cassandra
cluster = Cluster(['127.0.0.1'], port=9042)
session = cluster.connect()

# Sử dụng keyspace
session.set_keyspace('doanbigdata')

# Đọc dữ liệu từ file CSV
data = pd.read_csv('/home/tee/BIG_DATA/Air_quality_official.csv')

# Chèn dữ liệu vào bảng
insert_query = """
INSERT INTO air_quality (date, time, co_gt, pt08_s1_co, nmhc_gt, c6h6_gt, pt08_s2_nmhc, nox_gt, pt08_s3_nox, no2_gt, pt08_s4_no2, pt08_s5_o3, t, rh, ah)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

for index, row in data.iterrows():
    session.execute(insert_query, (
        row['Date'], row['Time'], row['CO(GT)'], row['PT08.S1(CO)'], row['NMHC(GT)'], row['C6H6(GT)'],
        row['PT08.S2(NMHC)'], row['NOx(GT)'], row['PT08.S3(NOx)'], row['NO2(GT)'], 
        row['PT08.S4(NO2)'], row['PT08.S5(O3)'], row['T'], row['RH'], row['AH']
    ))

print("Nhập dữ liệu thành công!")


    
# Đóng kết nối
cluster.shutdown()
```

```python
# Định nghĩa địa chỉ của node Cassandra
CASSANDRA_HOSTS = ['127.0.0.1']  # Thay bằng IP của node Cassandra (nếu có nhiều node, thêm các IP khác vào list)

# Tạo kết nối đến cluster
try:
    cluster = Cluster(CASSANDRA_HOSTS,port=9042)  # Khởi tạo cluster với danh sách node
    session = cluster.connect()        # Tạo session để thao tác với cơ sở dữ liệu
    
    # Chọn keyspace
    KEYSPACE_NAME = 'doanbigdata'  # Thay 'your_keyspace' bằng keyspace của bạn
    session.set_keyspace(KEYSPACE_NAME)
    
    print("Kết nối Cassandra thành công!")

    # Thực hiện một truy vấn mẫu
    query = "SELECT * FROM air_quality;"  # Thay 'your_table' bằng tên bảng của bạn
    rows = session.execute(query)
    
    print("Dữ liệu từ bảng:")
    for row in rows:
        print(row)
    
except Exception as e:
    print("Lỗi kết nối Cassandra:", str(e))

finally:
    # Đảm bảo đóng kết nối sau khi hoàn tất
    if 'cluster' in locals():
        cluster.shutdown()
```

```python
from cassandra.cluster import Cluster
```

```python
# Kết nối đến Cassandra
cluster = Cluster(['127.0.0.1'], port=9042)
session = cluster.connect()# Sử dụng keyspace
session.set_keyspace('test')


query = "SELECT * FROM air_quality WHERE date >= '01/09/2004' AND date <= '03/10/2004' ALLOW FILTERING;"  # Thay 'your_table' bằng tên bảng của bạn
rows = session.execute(query)
    
print("Dữ liệu từ bảng:")
for row in rows:
    print(row)
```

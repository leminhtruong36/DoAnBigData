```python
df2 = df1.drop(['Unnamed: 15', 'Unnamed: 16'], axis='columns')
df2.shape
```

```python
df2 = df2.replace(',', '.', regex=True)
df2.head(15)
```

```python
df2.isnull().sum()
```

```python
df3 = df2.fillna(0)
df3.isnull().sum()
```

```python
def replace(x):
    try:
        x = float(x)
        if x < 0:
            x = 0
        return x
    except ValueError:
        return np.nan
```

```python
columns_to_update = ['CO(GT)', 'PT08.S1(CO)', 'NMHC(GT)', 'C6H6(GT)', 
                     'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 
                     'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 
                     'T', 'RH', 'AH']

for column in columns_to_update:
    df3[column] = df3[column].apply(replace)
```

```python
import pandas as pd

# Đọc dữ liệu từ tệp CSV
file_path = "Air_quality_cleaned.csv"
df = pd.read_csv(file_path)

# Chuyển đổi cột Date sang định dạng YYYY-MM-DD
df["Date"] = pd.to_datetime(df["Date"], format="%d/%m/%Y").dt.strftime("%Y-%m-%d")

# Chuyển đổi cột Time sang định dạng HH:MM:SS
df["Time"] = pd.to_datetime(df["Time"], format="%H.%M.%S").dt.strftime("%H:%M:%S")

# Lưu lại dữ liệu đã định dạng
df.to_csv("Air_quality_official.csv", index=False)

```
